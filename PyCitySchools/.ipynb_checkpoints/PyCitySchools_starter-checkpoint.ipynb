### Observations based on Analysis done below

1. The school type matters. If you look at the analysis, all the top 5 schools are charter schools and the bottom 5 schools are district schools. Charter schools did better than district schools in math scores. It could be because these schools have a smaller student population per school.

2. District schools have a larger student population and a higher budget. They spend more money per student, however that did not lead to better scores. It could be due to district having a larger student population and maybe student-teacher ratio is greater in district schools, but more analysis and date is needed to see is this is due to such school practices.

3. Looking at sizes, small and medium schools performed way better on scores than large size schools (89-90% overall passing versus 58%) which can be seen from school type analysis as Charter schools are smaller than District schools.

```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left")
#school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
# count the number of rows in school_data
total_schools = len(school_data["school_name"].unique())
#total_schools

#Calculate the total number of students
total_students = len(student_data["student_name"])

#total_students
#Calculate the total budget
total_budget = school_data["budget"].sum()
#total_budget
```

```python
#Calculate the average math score
total_math_score = student_data["math_score"].sum()
average_math_score = total_math_score / total_students
average_math_score
#total_math_score


#Calculate the average reading score
total_read_score = student_data["reading_score"].sum()
average_read_score = total_read_score / total_students
#average_read_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_student = student_data.loc[student_data["math_score"] >= 70]
#math_student
count_math_student = math_student["student_name"].count()
percent_math_student = count_math_student/total_students * 100
#percent_math_student

#Calculate the percentage of students with a passing reading score (70 or greater)
read_student = student_data.loc[student_data["reading_score"] >= 70]
count_read_student = read_student["student_name"].count()
percent_read_student = count_read_student/total_students * 100
#percent_read_student

#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_student = student_data.loc[(student_data["reading_score"] >= 70) & (student_data["math_score"] >= 70)]
#overall_student
count_overall_student = overall_student["student_name"].count()
percent_overall_student = count_overall_student/total_students * 100
#percent_overall_student
```

```python
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({
    "Total Schools" : [total_schools],
    "Total Students" : [total_students],
    "Total Budget" : [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score" :[average_read_score],
    "% Passing Math" : [percent_math_student],
    "% Passing Reading" : [percent_read_student],
    "% Overall Passing" : [percent_overall_student]    
})

district_summary

#Optional: give the displayed data cleaner formatting

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.2f}".format)

district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math

* Create a dataframe to hold the above results

```python
#School Name
school_data_group = school_data_complete.groupby("school_name")
school_data_group.head()
#school_data

#School Type
school_type = school_data.set_index('school_name')['type']
#school_type.head()

#Total Students
total_school_student = school_data_group["student_name"].count()

#Total School Budget
school_budget = school_data.set_index('school_name')['budget']
#school_data_group['budget'].mean()
#school_budget.head()

#Per Student Budget
student_budget = school_budget/total_school_student
#student_budget

```

```python
#Average Math Score per school
avg_math = school_data_group["math_score"].mean()
#avg_math

#Average Reading Score
avg_read = school_data_group["reading_score"].mean()
#avg_read
```

```python
#% Passing Math
math_student = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")
#math_student
count_math_student = math_student["student_name"].count()
percent_math_student = count_math_student/total_school_student * 100
#percent_math_student

#% Passing Reading
read_student = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")
count_read_student = read_student["student_name"].count()
percent_read_student = count_read_student/total_school_student * 100
#percent_read_student

#% Overall Passing (The percentage of students that passed math and reading.)
overall_student = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("school_name")
count_overall_student = overall_student["student_name"].count()
percent_overall_student = count_overall_student/total_school_student * 100
#percent_overall_student
```

```python
#Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Type": school_type,
        "Total Students": total_school_student,
        "Total School Budget": school_budget,
        "Per Student Budget": student_budget,
        "Average Math Score": avg_math,
        "Average Reading Score": avg_read,
        "% Passing Math": percent_math_student,
        "% Passing Reading": percent_read_student,
        "% Overall Passing": percent_overall_student})

#formating
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df["Average Math Score"] = school_summary_df["Average Math Score"].map("{:,.2f}".format)
school_summary_df["Average Reading Score"] = school_summary_df["Average Reading Score"].map("{:,.2f}".format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map("{:,.2f}".format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("{:,.2f}".format)
school_summary_df["% Overall Passing"] = school_summary_df["% Overall Passing"].map("{:,.2f}".format)

school_summary_df
```

  * Group each series by school
  
  * Combine the series into a dataframe

```python
#Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement.
#Group each series by school
ninegrade = school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name")
ninegrade_math_avg = ninegrade["math_score"].mean()

tengrade = school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name")
tengrade_math_avg = tengrade["math_score"].mean()

elevengrade = school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name")
elevengrade_math_avg = elevengrade["math_score"].mean()

twelvegrade = school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name")
twelvegrade_math_avg = twelvegrade["math_score"].mean()

#Combine the series into a dataframe
math_avg_grade_df = pd.DataFrame({"9th" :ninegrade_math_avg,"10th":tengrade_math_avg,"11th" : elevengrade_math_avg,"12th" :twelvegrade_math_avg})

#Optional: give the displayed data cleaner formatting
math_avg_grade_df["9th"] = math_avg_grade_df["9th"].map("{:,.2f}".format)
math_avg_grade_df["10th"] = math_avg_grade_df["10th"].map("{:,.2f}".format)
math_avg_grade_df["11th"] = math_avg_grade_df["11th"].map("{:,.2f}".format)
math_avg_grade_df["12th"] = math_avg_grade_df["12th"].map("{:,.2f}".format)
math_avg_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Creating Data Frame for bin
bin_school_summary_df = pd.DataFrame({"School Type": school_type,
        "Total Students": total_school_student,
        "Total School Budget": school_budget,
        "Per Student Budget": student_budget,
        "Average Math Score": avg_math,
        "Average Reading Score": avg_read,
        "% Passing Math": percent_math_student,
        "% Passing Reading": percent_read_student,
        "% Overall Passing": percent_overall_student})

# created for later use in Scores by Size and Type
size_school_summary_df = bin_school_summary_df
type_school_summary_df = bin_school_summary_df

# Create the bins in which Data will be held
bins = [0,585,630,645,680]

# Create the names for the four bins
group_name = ["<$585", "$585-630","$630-645","$645-680"]


#Divide the data based on school spending
bin_school_summary_df['Spending Ranges(Per Student)'] = pd.cut(bin_school_summary_df["Per Student Budget"], bins, labels=group_name)

#bin_school_summary_df


```

```python
#group by average spending
spending_group = bin_school_summary_df.groupby('Spending Ranges(Per Student)')

#Average Math Score
spending_avg_math = spending_group["Average Math Score"].mean()
#spending_avg_math

#Average Reading Score
spending_avg_read = spending_group["Average Reading Score"].mean()
#spending_avg_read
```

```python
#% Passing Math
percent_math_student = spending_group["% Passing Math"].mean()
#percent_math_student

#% Passing Reading
percent_read_student = spending_group["% Passing Reading"].mean()
#percent_read_student

#Overall Passing Rate 
percent_overall_student = spending_group["% Overall Passing"].mean()
#percent_overall_student

```

```python
#Create dataframe
spending_score_df = pd.DataFrame({
    "Average Math Score": spending_avg_math,
    "Average Reading Score": spending_avg_read,
    '% Passing Math': percent_math_student,
    '% Passing Reading': percent_read_student,
    "Overall Passing": percent_overall_student})

#formating

spending_score_df["Average Math Score"] = spending_score_df["Average Math Score"].map("{:.2f}".format)
spending_score_df["Average Reading Score"] = spending_score_df["Average Reading Score"].map("{:.2f}".format)
spending_score_df["% Passing Math"] = spending_score_df["% Passing Math"].map("{:.2f}".format)
spending_score_df["% Passing Reading"] = spending_score_df["% Passing Reading"].map("{:.2f}".format)
spending_score_df["Overall Passing"] = spending_score_df["Overall Passing"].map("{:.2f}".format)


spending_score_df
```
